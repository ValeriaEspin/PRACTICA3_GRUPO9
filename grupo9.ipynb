LIMPIEZA, TRANSFORMACIÓN Y ADAPTACIÓN DE DATOS

```python
df.columns.tolist()  # Ver nombres originales
```

```python
df.rename(columns={
    'age': 'EDAD',
    'sex': 'GENERO',
    'cp': 'TIPO_DOLOR_PECHO',
    'trestbps': 'PRESION_SANGUINEA',
    'chol': 'COLESTEROL',
    'fbs': 'AZUCAR_AYUNAS',
    'restecg': 'RESULTADO_ELECTROCARDIOGRAMA',
    'thalach': 'FRECUENCIA_CARDIACA',
    'exang': 'EJERCICIO_ANGINA', 
    'oldpeak': 'CAMBIOS_ST_EJERCICIO',
    'slope': 'INCLINACION_ST_EJERCICIO',
    'ca': 'VASOS_CORONARIOS',
    'thal': 'FLUJO_CARDIACO',
    'condition': 'ENFERMEDAD_CORAZON'
}, inplace=True)
```

```python
df.head()
```

```python
df.describe()
```

```python
df['GENERO'] = df['GENERO'].replace({1: 'Masculino', 0: 'Femenino'})
```

```python
df['GENERO'].value_counts()
```

```python
df.head()
```

```python
#TIPO DE DOLOR EN EL PECHO
#Angina tipica: Dolor de pecho clásico asociado a esfuerzo físico o estrés, que mejora en reposo.
#Angina atípica: Dolor de pecho menos característico, puede aparecer en reposo o sin relación clara al esfuerzo.
#Dolor no anginoso: Dolor que no parece tener origen cardíaco (puede ser muscular, gástrico, etc.)
#Asintomático: La persona no presenta dolor en el pecho, pero puede tener enfermedad cardíaca detectada por otros medios (como pruebas de esfuerzo o ECG).

df['TIPO_DOLOR_PECHO'] = df['TIPO_DOLOR_PECHO'].replace({0: 'Angina tipica', 1: 'Angina atipica', 2: 'Dolor no anginoso', 3: 'Asintomatico'})
```

```python
#AZÚCAR EN AYUNAS, GLUCOSA - >120 RIESGO CARDIOVASCULAR
df['AZUCAR_AYUNAS'] = df['AZUCAR_AYUNAS'].replace({1: 'SI', 0: 'NO'})
```

```python
#ANOMALÍA: CAMBIOS EN EL CORAZÓN-PROBLEMA HIPERTRÓFIA: ENGROSAMIENTO DEL MÚSCULO DEL CORAZÓN - PRESIÓN ALTA U OTROS PROBLEMAS
df['RESULTADO_ELECTROCARDIOGRAMA'] = df['RESULTADO_ELECTROCARDIOGRAMA'].replace({0: 'Normal', 1: 'Anomalía ST-T', 2: 'Hipertrofia VI'})
```

```python
#DOLOR DE PECHO AL HACER EJERCICIO
df['EJERCICIO_ANGINA'] = df['EJERCICIO_ANGINA'].replace({1: 'SI', 0: 'NO'})
```

```python
#CAMBIOS_ST_EJERCICIO: CAMBIOS CARDIACOS EJERCICIO VS REPOSO
#INCLINACION_ST_EJERCICIO: 0: Ascendente (upsloping) – La línea del segmento ST sube después del ejercicio, lo cual suele ser un signo normal.
#1: Plana (flat) – La línea ST se mantiene recta; puede ser indicio de un posible problema cardíaco.
#2: Descendente (downsloping) – La línea ST baja; generalmente se asocia con mayor riesgo de enfermedad cardíaca.

df['INCLINACION_ST_EJERCICIO'] = df['INCLINACION_ST_EJERCICIO'].replace({0: 'Ascendente', 1: 'Plana', 2:'Descendente'})
```

```python
#FLUJO_CARDIACO
#0: Normal – Flujo sanguíneo normal al corazón.
#1: Defecto fijo – Parte del corazón no recibe suficiente sangre, incluso después del ejercicio (posible daño permanente).
#2: Defecto reversible – Parte del corazón no recibe suficiente sangre durante el ejercicio, pero mejora en reposo (posible isquemia).

df['FLUJO_CARDIACO'] = df['FLUJO_CARDIACO'].replace({0: 'Normal', 1: 'Defecto fijo', 2:'Defecto reversible'})
```

```python
#ENFERMEDAD_CORAZON

df['ENFERMEDAD_CORAZON'] = df['ENFERMEDAD_CORAZON'].replace({0: 'Sin enfermedad cardiaca', 1: 'Presencia de enfermedad cardiaca'})
```

```python
df.head()
```

```python
# Convertir variables a tipo 'category'
categorical_cols = ['EDAD', 'GENERO', 'TIPO_DOLOR_PECHO', 'PRESION_SANGUINEA', 'COLESTEROL', 'AZUCAR_AYUNAS', 'RESULTADO_ELECTROCARDIOGRAMA', 
'FRECUENCIA_CARDIACA', 'EJERCICIO_ANGINA', 'CAMBIOS_ST_EJERCICIO', 'INCLINACION_ST_EJERCICIO', 'VASOS_CORONARIOS', 'FLUJO_CARDIACO', 'ENFERMEDAD_CORAZON']

for col in categorical_cols:
    df[col] = df[col].astype('category')

```

ANÁLISIS EXPLOTARIO DE DATOS 

```python
#EDAD VS ENFERMEDAD_CORAZON

# Configurar estilo
sns.set(style="whitegrid")

# Crear el gráfico
plt.figure(figsize=(8, 6))
sns.violinplot(x='ENFERMEDAD_CORAZON', y='EDAD', data=df, palette='muted')

# Etiquetas y título
plt.xticks([0, 1], ['Con enfermedad', 'Sin enfermedad'])
plt.xlabel('Condición cardíaca')
plt.ylabel('Edad')
plt.title('Distribución de edad según presencia de enfermedad cardíaca')

plt.yticks(ticks=range(0, 91, 10))  # Mostrar ticks personalizados
plt.gca().invert_yaxis()

# Mostrar
plt.tight_layout()
plt.show()


```

```python
#GENERO VS ENFERMEDAD DEL CORAZON

# Crear gráfico de barras agrupadas
plt.figure(figsize=(8, 6))
ax = sns.countplot(data=df, x='GENERO', hue='ENFERMEDAD_CORAZON', palette='Set2')

for container in ax.containers:
    ax.bar_label(container, label_type='edge', padding=3, fontsize=10)

# Etiquetas y título
plt.xlabel('Sexo')
plt.ylabel('Cantidad de personas')
plt.title('Condición cardíaca según sexo')
plt.legend(title='Condición cardíaca')

plt.tight_layout()
plt.show()
```

```python
#COLESTEROL VS EDAD

# Crear gráfico de dispersión
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='COLESTEROL', y='EDAD', hue='ENFERMEDAD_CORAZON', palette='Set1')


# Etiquetas y título
plt.xlabel('Colesterol (mg/dl)')
plt.ylabel('Edad')
plt.title('Relación entre colesterol y edad')

plt.yticks(ticks=range(0, 91, 10))  # Mostrar ticks personalizados
plt.gca().invert_yaxis()

plt.tight_layout()
plt.show()
```

```python
#MAPA DE CALOR CORRELACIÓN ENTRE VARIABLES FISIOLÓGICAS

# Filtrar columnas numéricas
cols = ['EDAD', 'PRESION_SANGUINEA', 'FRECUENCIA_CARDIACA', 'CAMBIOS_ST_EJERCICIO']
corr = df[cols].corr()

# Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Mapa de correlación entre variables fisiológicas')
plt.show()

```
